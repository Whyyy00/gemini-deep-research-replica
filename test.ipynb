```python
from tavily import TavilyClient
import os
from dotenv import load_dotenv

load_dotenv()

tavily_client = TavilyClient(api_key=os.getenv("TAVILY_API_KEY"))

results = tavily_client.search(query="What is MCP?",
                               max_results=3,
                               topic="general",
                               include_raw_content="markdown")
```
